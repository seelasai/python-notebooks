# Understanding RDDs in Pyspark.

```python
# reading a rdd from text file
rdd = spark.sparkContext.textFile("/content/test.txt",4)
for element in rdd.collect():
  print(element)
```

```python
rdd.getNumPartitions()
```

### flatMap() Transformation:
flatMap() transformation flattens the RDD after applying the function and returns a new RDD. In other words, The flatMap() transformation is similar to the map() transformation but with one key difference: it allows each input element to map to zero or more output elements.

```python
rdd2 = rdd.flatMap(lambda x: x.split(' '))
rdd2.collect()
# splits the element by space and flattens the data.
```

###map()
map() transformation is used the apply any complex operations by applying a function to each element.

In our word count example, we are adding a new column with value 1 for each word, the result of the RDD is PairRDDFunctions which contains key-value pairs, word of type String as Key and 1 of type Int as value.

```python
rdd3= rdd2.map(lambda x: (x,1))
rdd3.collect()
```

###reduceByKey()
reduceByKey() merges the values for each key with the function specified. In our example, it reduces the word string by applying the sum function on value. The result of our RDD contains unique words and their count.

```python
rdd4 = rdd3.reduceByKey(lambda a,b: a+b)
```

```python
rdd4.count() #action command
```

###sortByKey()
sortByKey() transformation is used to sort RDD elements on key. In our example, first, we convert RDD[(String,Int]) to RDD[(Int,String]) using map transformation and later apply sortByKey which ideally does sort on an integer value. And finally, foreach with println statement prints all words in RDD and their count as key-value pair to console.

```python
rdd5 = rdd4.map(lambda x: (x[1],x[0])).sortByKey() #ascending=True
rdd5.collect()
```

###filter() Transformation.
filter() transformation is used to filter the records in an RDD. In our example we are filtering all words starts with “a”.


```python
rdd6 = rdd5.filter(lambda x: x[1].startswith("a"))
rdd6.collect()
```

```python
# prompt: generate a quarterly  stepup sip calculator
!apt-get install openjdk-8-jdk-headless -qq > /dev/null
!wget -q http://archive.apache.org/dist/spark/spark-3.1.2/spark-3.1.2-bin-hadoop3.2.tgz
!tar xf spark-3.1.2-bin-hadoop3.2.tgz
!pip install -q findspark
import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-8-openjdk-amd64"
os.environ["SPARK_HOME"] = "/content/spark-3.1.2-bin-hadoop3.2"
!wget -q https://raw.githubusercontent.com/spark-examples/spark-scala-examples/master/src/main/resources/test.txt
# some csv files, uncomment to download
#!wget -q https://sample-videos.com/csv/Sample-Spreadsheet-1000-rows.csv
#!wget -q https://www.stats.govt.nz/assets/Uploads/Business-operations-survey/Business-operations-survey-2022/Download-data/business-operations-survey-2022-information-and-communications-technology.csv
import findspark
findspark.init()
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("read_write").getOrCreate()
spark


# ### Defining the Quarterly Step-up SIP Calculator Function
def quarterly_stepup_sip_calculator(principal, rate, years, step_up_percentage):
  # Calculate the quarterly interest rate
  quarterly_rate = rate / 4

  # Calculate the number of quarters in the investment period
  num_quarters = years * 4

  # Initialize the list to store the quarterly balances
  quarterly_balances = [principal]

  # Loop through each quarter
  for i in range(1, num_quarters):
    # Calculate the interest earned in the current quarter
    interest = quarterly_balances[i-1] * quarterly_rate

    # Apply the step-up percentage to the principal
    if i % 4 == 0:
      principal *= (1 + step_up_percentage)

    # Calculate the new quarterly balance
    quarterly_balances.append(quarterly_balances[i-1] + interest + principal)

  # Return the list of quarterly balances
  return quarterly_balances

# ### Sample Usage
principal = 5000
rate = 50
years = 20
step_up_percentage = 10

quarterly_balances = quarterly_stepup_sip_calculator(principal, rate, years, step_up_percentage)

# Print the quarterly balances
for i, balance in enumerate(quarterly_balances):
  print(f"Quarter {i+1}: {balance}")
```
